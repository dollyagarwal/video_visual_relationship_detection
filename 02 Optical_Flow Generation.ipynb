```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def optical_flow(one, two):
    one_g = cv2.cvtColor(one, cv2.COLOR_BGR2GRAY)
    two_g = cv2.cvtColor(two, cv2.COLOR_BGR2GRAY)
    hsv = np.zeros_like(one)
    # set saturation
    hsv[:,:,1] = cv2.cvtColor(two, cv2.COLOR_RGB2HSV)[:,:,1]
    # obtain dense optical flow paramters
    flow = cv2.calcOpticalFlowFarneback(one_g, two_g, None,0.5, 3, 15, 3, 5, 1.2, 0)
    # convert from cartesian to polar
    mag, ang = cv2.cartToPolar(flow[..., 0], flow[..., 1])
    # hue corresponds to direction
    hsv[:,:,0] = ang * (180/ np.pi / 2)
    # value corresponds to magnitude
    hsv[:,:,2] = cv2.normalize(mag,None,0,255,cv2.NORM_MINMAX)
    # convert HSV to int32's
    # hsv = np.asarray(hsv, dtype= np.float32)
    rgb_flow = cv2.cvtColor(hsv,cv2.COLOR_HSV2RGB)
    return rgb_flow 
```

```python
import os
data_dir = 'Data/augmented_images'
optical_dir = 'Data/optical_flow_images/train'
for foldername in os.listdir(data_dir):
  
    list_images = sorted(os.listdir(data_dir+'/'+foldername))
    os.makedirs(optical_dir + '/' + foldername, exist_ok=True) 
    for i in range(0, 30):
      if i==0:
        frame1 = list_images[i]
        frame2 = list_images[i]
      else:
        frame1 = list_images[i-1]
        frame2 = list_images[i]   
      path1 = os.path.join(data_dir, foldername, frame1)
      path2 = os.path.join(data_dir, foldername, frame2)
      rgb = optical_flow(cv2.imread(path1), cv2.imread(path2))  
      cv2.imwrite(os.path.join(optical_dir, foldername, frame2), rgb)
    print(foldername)
```
