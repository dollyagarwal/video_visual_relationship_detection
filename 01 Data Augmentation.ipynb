```python
rotationAngle = 15
scale_out_factor = 1.1
transform_x = 25
transform_y = 25
noise_factor = 0.155
numAug = 1
```

```python
from skimage import *
```

```python
import numpy as np
from skimage.io import imread, imsave
import matplotlib.pyplot as plt
from skimage import transform
from skimage.transform import rotate, AffineTransform,warp,rescale
from skimage.util import random_noise
from skimage.filters import gaussian
from scipy import ndimage
from skimage import exposure
```

```python
def augment(img, methodsToUse):
  if 'rotate' in methodsToUse:
    img = rotate(img, angle=rotationAngle, mode = 'constant', cval = 0)
  if 'scale_out' in methodsToUse:
     img = rescale(img, scale=scale_out_factor)  
  if 'transform'  in methodsToUse:
     transform = AffineTransform(translation=(transform_x,transform_y))
     img = warp(img,transform,mode='wrap')  
  if 'noise' in methodsToUse:
     img = random_noise(img,var=noise_factor**2)  
  if 'blur' in methodsToUse:
     img = gaussian(img,sigma=1,multichannel=True)
  if 'bright' in methodsToUse:
    img = exposure.adjust_gamma(img, 1)
  if 'contrast' in methodsToUse:
    img = exposure.adjust_gamma(img, 2) #img = img * 1.5
  return img  
```

# Augmentation

```python
import pandas as pd
folders=pd.read_csv('Data/unpopular_videos.csv')
```

```python
##Count of images to augment according to the augmented data study

dict={6: 1,
16: 4,
7: 7,
22: 9,
3: 9,
17: 12,
14: 5,
19: 9,
31: 7,
27: 11,
4: 10,
28: 10}
print(dict[6])
```

```python
for k in dict.keys():
    samples = folders[(folders['object1']==k) | (folders['object2']==k)].sample(n=dict[k],replace=True)
    c=0
    print('key',k)
    for i,folder in samples.iterrows():
      c+=1
      #print('folder',folder.path)

      methods = ['rotate','scale_out','transform','noise','blur','bright','contrast']
      probabilities = np.random.random(len(methods))
      methodsToUse = []
      for idx, val in enumerate(probabilities):
        if val > 0.5:
          methodsToUse.append(methods[idx])
      #print(methodsToUse)
      f = folder.path.split('/')[-1]
      p = os.path.join(output_dir,f+'-'+str(k)+'-'+str(c))
      os.mkdir(p)
      for images in sorted(os.listdir(folder['path'])):
        #print(images)
        image = imread(folder['path']+'/'+images) / 255
        #print('image shape',image.shape)
        plt.imshow(image)
        plt.show()
        augmentedImage = augment(image, methodsToUse)
        #print(augmentedImage.shape)
        plt.imshow(augmentedImage)
        plt.show()
        output_frame_path=os.path.join(p,images)
        #print(output_frame_path)
        plt.imsave(output_frame_path, augmentedImage)
        #cv2.imwrite(output_frame_path, augmentedImage)     
```
