```python
complete_dic ={}
for i,x in data.iterrows():
    if x['object1'] in complete_dic:
      complete_dic[x['object1']]+=1
    else:
      complete_dic[x['object1']]=1
    if x['object2'] in complete_dic:
      complete_dic[x['object2']]+=1
    else:
      complete_dic[x['object2']]=1

```

```python
print('Total Classes',len(complete_dic))
print('Classwise Counts',complete_dic)
```

```python
unpopular_classes = {k : v for k,v in complete_dic.items() if v <= 12}
```

```python
print('Total Unpopular Classes',len(unpopular_classes))
print('Unpopular class counts',unpopular_classes)
```

```python
count1, count2, count3, count4 = 0,0,0,0
ind=[]
popular,unpopular=[],[]
for i,r in data.iterrows():
  if r['object1'] in unpopular_classes and r['object2'] in unpopular_classes:
    count1+=1
    ind.append(i)
    unpopular.append(i)
  if r['object1'] in unpopular_classes and r['object2'] not in unpopular_classes:
    count2+=1
    ind.append(i)
    unpopular.append(i)
  if r['object1'] not in unpopular_classes and r['object2'] in unpopular_classes:
    count3+=1
    unpopular.append(i)
    ind.append(i)
  if r['object1'] not in unpopular_classes and r['object2'] not in unpopular_classes:
    count4+=1
    popular.append(i)

print('Videos containing both unpopular objects',count1)
print('Videos containing any one unpopular objects',count2+count3)
print('Videos containing both popular objects',count4)
```

```python
print('Len:',len(unpopular), 'Indices of videos with atleast one unpopular object',unpopular)
```

```python
print('Len:',len(popular), 'Indices of videos with with both popular object',popular)
```

```python
popularDataFrame=data.iloc[popular]
unpopularDataFrame=data.iloc[unpopular]
```

```python
unpopularDataFrame.head()
```

```python
count = 0
ind = []
for i,r in unpopularDataFrame.iterrows():
  if r['object1']  in unpopular_classes and r['object2']  in unpopular_classes:
    count+=1
    ind.append(i)
print('Count of videos where both objects are unpopular',count)
print('Indices of videos where both objects are unpopular',ind)
```

```python
unpopularDataFrame[unpopularDataFrame.index.isin(ind)].to_csv('{}/unpopular_videos.csv'.format(location),index=False)
```
