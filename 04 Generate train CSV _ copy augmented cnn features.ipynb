# Copy Augmented cnn features to train cnn features folder

```python
import os
import shutil

aug_location = 'augmented_cnn'
train_location = 'train_cnn'
new = train
c=447
for file in sorted(os.listdir(aug_location)):
  filename = file.split('_')[-1]
  f = filename.split('.')[0]
  object1 = data[f][0]
  relation = data[f][1]
  object2 = data[f][2]
  new = new.append({'index':'f-{}'.format(file),'object1':object1,'relation':relation,'object2':object2},ignore_index=True)
  shutil.copyfile(os.path.join(aug_location,file),os.path.join(train_location,'train_cnnfeatures{}.npy'.format(c)))
  c+=1
```

```python
new.to_csv('train.csv',index=False)
```
